```python
balanced_data = []
for org, group in df.groupby('org_code'):
    unique_group = group.drop_duplicates() 
    if len(unique_group) >= 2000:
        group_balanced = unique_group.sample(n=2000, random_state=42)
    else:
        remaining = 2000 - len(unique_group)
        extra = unique_group.sample(n=remaining, replace=True, random_state=42)
        group_balanced = pd.concat([unique_group, extra])
    balanced_data.append(group_balanced)
balanced_df = pd.concat(balanced_data)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
english_stopwords = set(stopwords.words('english'))
```

```python
hindi_stopwords = {
'और','का','की','के','है','हैं','को','से','पर','यह','था','थे','कर','भी',
'में','तो','ही','लेकिन','अगर','जब','तक','क्यों','कौन','क्या'
}
```

```python
all_stopwords = english_stopwords.union(hindi_stopwords)
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\u0900-\u097F\s]', '', text)
    words = text.split()
    cleaned_words = []
    for word in words:
        if word not in all_stopwords:
            if re.match(r'[a-zA-Z]+$', word):
                word = lemmatizer.lemmatize(word)
            cleaned_words.append(word)
    return " ".join(cleaned_words)
```

```python
balanced_df['subject_content_text'].apply(clean_text)
```
